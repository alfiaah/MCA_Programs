```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x = [5,15,25,35,45,55];
y = [5,20,14,32,22,30];

#Linear Regression

plt.scatter(x,y)
plt.show()
```

```python
slope,intercept,r,p,std_err=stats.linregress(x,y)

def myfunc(x):
    return slope*x+intercept

mymodel=list(map(myfunc,x))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```

```python
print(r)
print(p)
```

```python
#Logistic Regression
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x = [1,2,3,5,6,7,8,10,12,13,14,15,16,18,19,20]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99]

# Use np.poly1d to create a polynomial object from the coefficients
mymodel = np.poly1d(np.polyfit(x,y,3))

myline = np.linspace(1,20,100)

plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_score(y,mymodel(x))
```
