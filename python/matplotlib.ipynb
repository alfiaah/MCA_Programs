```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Line Plot
x=[1,2,3,4]
y=[11,22,33,44]

plt.plot(x,y)
plt.show()
```

```python
#Line Plot
x=[1,2,3,4]
y=[11,22,33,44]

plt.plot(x,y,color='red',linestyle='-',marker='o',label='Line Plot')
plt.legend()
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Scatter Plot

import matplotlib.pyplot as plt
import numpy as np

x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,97,94,78,92,56,76]

plt.scatter(x,y,color='powderblue',marker='x',label='Scatter Plot')
plt.legend()
plt.title('Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Bar Plot

import matplotlib.pyplot as plt
import numpy as np

categories=['A','B','C','D']
values=[1,4,5,2]

plt.bar(categories,values,color='green',label='Bar Plot')

plt.show()
```

```python
#Bar Plot Horizontal

import matplotlib.pyplot as plt
import numpy as np

categories=['A','B','C','D']
values=[1,4,5,2]

plt.barh(categories,values,color='green',label='Bar Plot')

plt.show()
```

```python
#Histogram

import matplotlib.pyplot as plt

data=[4,6,2,5,6,3,10,8,5]

plt.hist(data,bins=5,color='red',label='Histogram')

plt.show()
```

```python
#Pie Chart

labels=['Apple','Banana','Mango','Orange'];
sizes=[215,130,245,200];
colors=['red','yellow','green','orange'];
explode = (0,0,0,0.2)
plt.title('Pie Chart')
plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct='%1.1f',shadow=True,startangle=90);
plt.axis('equal');
plt.show();
```

```python
#Box Plot
import matplotlib.pyplot as plt
import numpy as np

data=[2,3,5,8,12],[3,4,7,10,14],[0,2,6,8,13]
plt.boxplot(data, vert=True, patch_artist=True, labels=['A','B','C'])
plt.title('Box Plot')
plt.show();
```

```python
#Violin Plot

data=[2,3,5,8,12],[3,4,7,10,14],[0,2,6,8,13]
plt.violinplot(data,vert=True,showmeans=True,showmedians=True)
plt.show()
```
